```python
stu1=[56,78,92,95,100]
stu2=[78,36,89,98,67]
sub=["tel","hin","eng","mat","sci"]
plt.plot(sub,stu1,label="student1")
plt.plot(sub,stu2,label="student2")
plt.title("MARKS OF STUDENTS")
plt.legend()
```

```python
stu1=[56,78,92,95,100]
stu2=[78,36,89,98,67]
sub=["tel","hin","eng","mat","sci"]
plt.subplot(2,1,1)
plt.plot(sub,stu1,label="student1",color="b")
plt.legend()
plt.subplot(2,1,2)
plt.plot(sub,stu2,label="student2",color="r")
plt.title("MARKS OF STUDENTS")
plt.legend()
```

```python
year=["19-20","20-21","21-22","22-23"]
a=np.array([230,560,780,127,128])
b=np.array([200,160,270,127,400])
proa=np.diff(a)
prob=np.diff(b)
print("Revenue of company A ",a)
print("Revenue of company B ",b)
print("Profits of company A ",proa)
print("Revenue of company B ",prob)
plt.subplot(1,2,1)
plt.bar(year,proa,label="proA",color="black")
plt.legend(loc="best")
plt.title("profits of companies A and B")
plt.subplot(1,2,2)
plt.bar(year,prob,label="proB",color="green")
plt.legend(loc="best")
```
